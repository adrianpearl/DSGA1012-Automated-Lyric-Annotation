```python
from langdetect import detect
from langdetect import detect_langs
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('annotations.tsv',sep='\t',header=None,names=['song_id','lyric','annotation'])
```

```python
df.head()
```

```python
lang_dict = {}
sample_size = 100
for song_id in df.song_id.unique()[:sample_size]:
    song = df[df.song_id == song_id]
    song['lang'] = song['lyric'].map(detect)
    percent = np.average((song['lang'] == 'en').astype(int))
    lang_dict[song_id] = percent
    print(song[['song_id','lang']])
```
